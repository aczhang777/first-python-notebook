What if we wanted to merge out two datasets together? The contribs data has a list of all contributions to committees, and props has info on whether those committees support or oppose certain propositions. It makes sense to merge these together.

We'll merge using committee IDs.

```python
merged = pd.merge(prop, contribs, on = "calaccess_committee_id") 
## the on parameter specifies which column you're using to merge. Also note that we're merging prop with contribs, so it's just info on prop 64!
```

```python
merged.head()
```

```python
merged.info()
```

```python
merged.amount
```

How much money was donated to committees on either side of prop 64?

```python
merged.amount.sum()
```

```python
merged.committee_position.value_counts() ## number of unique values in committee position
```

Let's separate out support from oppose.

```python
support = merged[merged.committee_position == 'SUPPORT']
```

```python
oppose = merged[merged.committee_position == 'OPPOSE']
```

```python
support.info()
```

```python
oppose.info()
```

```python
support.amount.sum() ## gives us the total amount of money given in support of prop 64
```

```python
oppose.amount.sum() ## gives us the total amount given in opposition of 64
```

```python
support.amount.sum() / merged.amount.sum() ## what fraction of total amount of money given was given in support?
```

```python
merged.sort_values("amount") ## this sorts the values in amount in ascneding order
```

```python
merged.sort_values("amount", ascending = False) ## this is how to sort in descending order
```

```python
merged.groupby('committee_position')
```

This output tells you that the output can't be neatly printed, but it does exist. It is not an error message.

```python
merged.groupby('committee_position').amount.sum()
```

Got the same numbers as above: the sum of money given to both sides.

```python
merged.groupby('committee_name_x').amount.sum().reset_index().sort_values("amount", ascending = False)
## we group our data by commitee name (the x means it's the committee names on the left side of merged, because
## when merging we keep all variable names, even duplicates, so committee_name_x is the one on the left), then take 
## the amount variable, sum it, print it as a table, sort the values in descending order
```

What if we wanted to group by 2 different things: first name and last name?

```python
merged.groupby(['contributor_firstname', 'contributor_lastname']).amount.sum().reset_index().sort_values("amount", ascending = False) 
## we put the two fields in brackets, separated by a comma
```

Corporations also donated but did not have a first name. We must correct for this.

```python
merged.contributor_firstname
```

```python
merged.contributor_firstname.fillna('***') ## we can fill in the NAs with whatever we want in its place
```

```python
merged.contributor_firstname = merged.contributor_firstname.fillna('***')
```

Another way of doing this: merged.contributor_firstname.fillna('***', inplace = TRUE)

```python
merged.groupby(['contributor_firstname', 'contributor_lastname']).amount.sum().reset_index().sort_values("amount", ascending = False)
```

Now this also shows info on corporations. Yay!

Note: when grouping by multiple items, you must pass in those items as a list because the input to groupby must be a single object. That's why in these examples we wrote ['contributor_firstname', 'contributor_lastname'].
